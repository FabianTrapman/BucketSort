```python
def run_cpp_program():
    result = subprocess.run(["./my_cpp_program.exe"], stdout=subprocess.PIPE, text=True)
    return result.stdout
```

```python
print(run_cpp_program())
```

Overigens is er gebruikt gemaakt van object oriÃ«nted programming.

```python
def analyze_timing():
    output = run_cpp_program()
    timing_values = [tuple(map(int, pair.split(', '))) for pair in output.split('\n') if pair]

    print(timing_values)

    # Unpack the tuples into separate lists for x and y values
    x_values, y_values = zip(*timing_values)

    # Plot the data
    plt.plot(x_values, y_values, marker='o', linestyle='-')
    plt.title('Timing Analysis')
    plt.xlabel('X Values')
    plt.ylabel('Y Values')
    plt.show()
```
